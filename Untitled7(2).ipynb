```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def number_1():
    cols = ['Name', 'Gender','Count']
    names_of_years = {}
    for year in range(1900, 2000):
        names_of_years[year] = pd.read_csv('yob'+str(year)+'.txt', names=cols)

    pd.concat(names_of_years, names=['Year']).groupby([pd.concat(names_of_years, names=['Year']).index.get_level_values(0), 'Name']).sum().query('Name == ["Ruth","Robert"]').unstack('Name').plot()
```

```python
number_1()
```

```python
def number_of_letters(string):
    LIST = ['b','c','d','f','g','h','j','k','l','m','n','p','q','r','s','t','v','w','x','z']
    string = string.lower()
    result = 0
    for x in string:
        if x in LIST:
            result += 1
    return result
```

```python
cols = ['Name','Gender','Count']
names_of_years = {}
for year in range(1900, 2000):
    names_of_years[year] = pd.read_csv('yob'+str(year)+'.txt', names=cols)
    names = pd.concat(names_of_years, names=['Year','Pos'])
sum_of_names = names.groupby(['Name','Gender']).sum()
count_letters = sum_of_names.index.get_level_values('Name').map(number_of_letters)
sum_of_names.assign(count_letters=count_letters).plot(kind = "scatter", x='count_letters', y='Count')
```
